### Conjuntos
<p>
É uma coleção não ordenada de objetos.

### Elementos de um conjunto
<p>
Os objetos no conjunto são chamados de elementos, ou membros, do conjunto. Diz-se que os elementos pertencem ao conjunto. </p>
<br/>
<ul>
    <li> $a \in A$: indica que $a$ é um elemento do conjunto A </li>
    <li> $a \not\in A$: indica que $a$ não é um elemento do conjunto A </li>
</ul>

### Exemplo
O conjunto $A$ das vogais do alfabeto português pode ser escrito por extenso:
$$
A= \{a,e,i,o,u \}
$$
ou por construção, considerando o domínio as letras do alfabeto português:
$$
A= \{ x| \; x \text{ é uma vogal} \}
$$

### Exemplo
O conjunto $B$ dos inteiros positivos menores que 100, pode ser escrito por extenso com reticências:
$$
B = \{ 1,2,3, \cdots, 98,99  \}
$$
ou por construção:
$$
B= \{ x \in \mathcal{Z}^+ | \; 1 \leq x<100   \}
$$

### Igualdade de Conjuntos
<p> Dois conjuntos são iguais se, e somente se, eles tem os mesmos elementos. </p>
<br/>
$$
A=B \longleftrightarrow \forall x \; (x \in A \leftrightarrow x \in B)
$$

### Exemplo
<p> Os conjuntos </p>
<center> $A=\{2, \; 4, \; 6  \}$ e $B=\{4, \; 6, \;  2 \}$ </center>
<p> são iguais pois tem os mesmos elementos </p>
<p> A ordem em que são listados os elementos não é relevante. </p>

### Exemplo
<p> Os conjuntos</p>
<center>$A=\{2, \; 4, \; 6  \}$ e $B=\{4, \; 4, \; 4, \; 6, \;  6, \; 2, \; 2, \; 2, \; 2 \}$ </center>
<p>são iguais pois tem os mesmos elementos. </p> 
<p> Não é relevante se um elemento for listado mais de uma vez. </p>

### Conjunto vazio
<p> O conjunto vazio, ou conjunto nulo, é o conjunto que não tem elementos. </p>
<p>É denotado por o símbolo $\Phi$ ou por um conjunto demarcado por um par de chaves $\{ {} \}$. </p>

### Exemplo
<p> O conjunto dos números reais $x$ que verificam </p>
<center> $x^2=-1$ </center>
<p> é o conjunto nulo, pois o quadrado de qualquer número real é não negativo. </p>

### Exemplo
<p> O conjunto de municipios do Estado do Amapá com mais de um milhão de habitantes é o conjunto nulo. </p>
<p> A população de todo o estado do Amapá em 2017 é estimada em 797.722 pessoas. </p>

### Conjunto Unitário
<p>Um conjunto unitário é um conjunto com um único elemento. </p>

### Exemplo
<p> O conjunto de números reais que verificam a equação </p>
<center> $x^3=1$ </center>
<p> é um conjunto unitário, pois $x=1$ é o único número real que verifica a equação. </p>

### Exemplo
<p> As cidades do estado do Amapá com população estimada em 2017 acima de 400.000 habitantes é um conjunto unitário, pois Macapá é a única cidade do estado do Amapá com tal condição. <p>

### Subconjunto
<p> Um conjunto $A$ é um subconjunto , denotado $A \subset B$,  se, e somente se, todo elemento de $A$ é um elemento de $B$. </p>
<br/>
\[
A \subset B \longleftrightarrow \forall x \; (x \in A \rightarrow x \in B)
\]

### Propriedades
<p> Para todo conjunto $A$: </p>
<ul>
    <li> $\Phi \subset A$ </li>
<br/>
    <li> $A \subset A$ </li>
</ul>

### Demonstração $\Phi \subset A$
$$
\forall x \; (\underbrace{\underbrace{x \in \Phi}_{falso} \rightarrow x \in A}_{verdadeiro})
$$
<p> portanto, $\Phi \subset A$. </p>

### Demonstração $A \subset A$
$$
\forall x \; (\underbrace{x \in A \rightarrow x \in A}_{verdadeiro})
$$
<p> portanto, $A \subset A$. </p>

### Propriedade
$$
A=B \longleftrightarrow ((A \subset B) \wedge (B \subset A))
$$

#### Prova de $A=B \longrightarrow ((A \subset B) \wedge (B \subset A))$
Se $A=B$,
<ul>
    <li> Como $A \subset \underbrace{A}_{=B}$, temos $A \subset B$ </li>
<li> Como $\underbrace{A}_{=B} \subset A$, temos $B \subset A$ </li>
</ul>
<p> logo, $(A \subset B) \wedge (B \subset A)$. </p>

#### Prova de $ ((A \subset B) \wedge (B \subset A)) \longrightarrow  A=B$
Se $(A \subset B) \wedge (B \subset A)$,
<ul>
<li> se $A \subset B$, então $\forall x \; (x \in A \rightarrow x \in B)$ </li>
<li> se $B \subset A$, então 
$\forall x \; (x \in B \rightarrow x \in A)$ </li> 
</ul>
<p> logo, $\forall x \; ((x \in A \rightarrow x \in B) \wedge 
 (x \in B \rightarrow x \in A))$, pelo qual </p>
\[
\forall x \; (x \in A \leftrightarrow x \in B)
\]
e portanto, $A=B$.

### Conjunto finito
<p> $A$ é um conjunto finito se contém exatamente $m$ elementos distintos. </p> 
<br/>
<p> O número $m$ é o cardinal de $A$, sendo denotado por $n(A)$ ou  $|A|$. </p>

#### Exemplos
<ul>
<li> Se $A$ é o conjunto das letras do alfabeto português, ele é um conjunto finito e $n(A) =26$. </li>
<br/>
<li> Se $\Phi$  é conjunto vazio, então ele não tem elementos. Portanto, o conjunto vazio é finito e $n(\Phi) =0$. </li>
    </ul>

### Conjunto infinito
<p> Um conjunto é infinito, se ele não é finito. </p>

#### Exemplo
<p> O conjunto dos números inteiros positivos é infinito. </p>

### Conjunto de Partes
<p> O conjunto de partes de um conjunto $A$, denotado por $\mathcal{P}(A)$, é o conjunto formado por todos os subconjuntos de $A$. </p>

#### Exemplo
$$
\mathcal{P}(\{a,b,c\}) = \{ \Phi, \; \{ a\},  \{b\}, \{c\}, \{a,b\}, \{a,c\}, \{b,c\}, \{a,b,c\}   \}
$$

#### Exemplo
<p> O conjunto de partes do conjunto vazio é: </p>
$$
\mathcal{P}(\Phi) = \{ \Phi \}
$$

#### Exemplo
<p> O conjunto de partes do conjunto formado pelo conjunto vazio é: </p>
$$
\mathcal{P}(\{ \Phi \}) = \{ \Phi, \{ \Phi \}  \}
$$

### Propriedade
<p> O cardinal de um conjunto finito de $n$ elementos é $2^n$. </p>

#### Demonstração
<p> Cada elemento do conjunto tem duas possibilidades: pode ou não ser escolhido para formar parte de um subconjunto, logo temos </p>
$$
\underbrace{2 \cdot 2 \cdots 2}_{n \; fatores} = 2^n
$$
<p> possibilidades de formar subconjuntos. </p>

### Par ordenado
<p> O par ordenado $(a_1,a_2)$ é a coleção ordenada que tem $a_1$ como primeiro elemento e $a_2$ como segundo elemento. </p>
<ul>
<li> <p> Observe que os conjuntos $\{a_1,a_2\}$ e $\{a_2, a_1\}$ são iguais por terem os mesmos elementos. </p>
    <p> Em conjuntos a ordem dos elementos não é relevante. </p> </li>
    <br/>
    <li> <p> Os pares ordenados $(a_1,a_2)$ e $(a_2,a_1)$ são diferentes, a menos que $a_1=a_2$. </p>
        <p> A ordem nos pares ordenados é relevante. </p> </li>
</ul>

### $n$-upla ordenada
<p> A n-upla ordenada $(a_1,a_2, \cdots,a_n)$ é a coleção ordenada que tem $a_1$ como seu primeiro elemento, $a_2$ como seu segundo elemento, ... e $a_n$ como seu n-ésimo elemento. </p>

### Produto Cartesiano
<p> O produto cartesiano de $A$ e $B$, denotado por $A \times B$, é o conjunto de todos os pares ordenados $(a,b)$, em que $a \in A$ e $b \in B$. Isto é, </p>
<br/>
$$
A \times B = \{ (a,b) \mid a \in A \wedge b \in B \}
$$

#### Exemplo
<p> Qual é o produto cartesiano $A \times B$, onde $A= \{a,b\}$ e $B= \{1,2,3\}$. </p>

```python
import numpy as np
def cartesiano(A,B):
    C=set()
    for x in A:
        for y in B:
            C.add((x,y))
    return(C)
```

```python
A={'a','b'}
B={1,2,3}
C=cartesiano(A,B)
print('AxB = ',C)
D=cartesiano(B,A)
print('BxA = ',D)
print('n(AxB) = ', len(C))
print('AxB = BxA?',C==D)
```

#### Solução
<p> $ A \times B = \{(a,1),(a,2),(a,3),(b,1),(b,2),(b,3)  \} $ </p>
<p> enquanto $B \times A = \{(1,a),(1,b),(2,a),(2,b),(3,a),(3,b)  \} $ </p>
<p> Assim em geral, </p>
$$
A \times B \neq B \times A.
$$
Observe que,
$$
n(A \times B) = n(A) \cdot n(B)
$$
<p> no exemplo, $n(A)= 2$, $\; n(B) = 3 \; $ e $ \; n(A \times B) = 2 \cdot 3 =6$. </p>

# <center> OPERAÇÕES COM CONJUNTOS </center>

### União
<p> A união dos conjuntos $A$ e $B$, denotada por $A \cup B$, é o conjunto que contém aqueles elementos que estão em $A$ ou em $B$, ou em ambos. </p>
<br/>
$$
A \cup B = \{ x \mid x \in A \vee x \in B \}
$$



### Interseção
<p> A interseção dos conjuntos $A$ e $B$, denotada por $A \cap B$, é o conjunto que contém aqueles elementos que estão em $A$ e em $B$, simultaneamente. </p>
$$
A \cap B = \{ x \mid x \in A \wedge x \in B \}
$$



### Conjuntos disjuntos
<p> Os conjuntos $A$ e $B$ são disjuntos se, e somente se, sua interseção é vazia, isto é $A \cap B = \Phi$. </p>

### Princípio de inclusão-exclusão
<ul>
    <li> $n(A \cup B) = n(A) + n(B), \quad \text{se  } A \cap B = \Phi$. </li>
    <br/>
    <li> $n(A \cup B) = n(A) + n(B) - n(A \cap B) $. </li> 
</ul>

### Diferença
<p> A diferença dos conjuntos $A$ e $B$, denotado por $A - B$, é o conjunto que contém os elementos que estão em $A$ mas não estão em $B$. Isto é, </p>
$$
A -B = \{ x \mid x \in A \wedge x \not\in B  \}.
$$



### Complemento
<p> Seja $U$ o conjunto universo. O <b> complemento </b> do conjunto $A$, denotado por $A^c$ ou $\bar{A}$, é a diferença de $A$ em relação a $U$, isto é: </p>
\begin{align}
A^c &= U-A \\
A^c &= \{x \mid x \not\in A  \}
\end{align}



### Elementos Neutros
 <ul>
    <li> <p> $   A \cup \Phi = A$ </p> </li>
    <br/>
    <li> <p> $  A \cap U = A$ </p> </li>
 </ul>

#### Demonstração
\begin{align}
A \cup \emptyset &= \{
x \mid x \in A \vee \underbrace{x \in \emptyset}_{falso}
\} \\
&= \{
x \mid x \in A
\} \\
&= A \\
& \\
A \cap U &= \{
x \mid x \in A \wedge \underbrace{x \in U}_{verdadeiro}
\} \\
&= \{
x \mid x \in A
\} \\
&= A
\end{align}

### Dominação
<ul>
    <li> <p> $   A \cup U = U$ </p> </li>
    <br/>
    <li> <p> $  A \cap \Phi = \Phi $ </p> </li>
</ul>

#### Demonstração
\begin{align}
A \cup U &= \{
x \mid x \in A \vee x \in U
\} \\
&= \{
x \mid x \in U \vee x\in U
\} \\
&= \{
x \mid x \in U
\} \\
&= U \\
& \\
A \cap \emptyset &= \{
x \mid \underbrace{x \in A \wedge \underbrace{x \in \emptyset}_{falso}}_{falso}
\} \\
&= \emptyset
\end{align}

### Idempotentes 
<ul>
    <li> $A \cup A  = A$ </li>
<br/>
    <li> $A \cap A = A$ </li>
</ul>

### Complementação
<ul>
    <li> $(A^c)^c = A$ </li>
</ul>


#### Demonstração
\begin{align}
(A^c)^c &= \{
x \mid x \in (A^c)^c
\} \\
&= \{
x \mid \neg (x \in A^c)
\} \\
&= \{
x \mid \neg \neg (x \in A)
\} \\
&= \{
x \mid x \in A
\} \\
&= A
\end{align}

### Comutativas
<ul>
    <li> $A \cup B =  B \cup A$ </li>
<br/>
    <li> $A \cap B = B \cap A$ </li>
</ul>

#### Demonstração
\begin{align}
A \cup B &= \{
x \mid x \in A \vee x \in B
\} \\
&= \{
x \mid x \in B \vee x \in A
\} \\
&= B \cup A \\
& \\
A \cap B &= \{
x \mid x \in A \wedge x \in B
\} \\
&= \{
x \mid x \in B \wedge x \in A
\} \\
&= B \cap A
\end{align}

### Associativas
<ul>
    <li> $A \cup (B \cup C) = (A \cup B) \cup C$ </li>
<br/>
    <li> $A \cap (B \cap C) = (A \cap B) \cap C$ </li>
</ul>

#### Demonstração
\begin{align*}
A \cup (B \cup C) &= \{
x \mid x \in A \vee x \in (B \cup C)
\} \\
&= \{
x \mid x \in A \vee (x \in B \vee x \in C)
\} \\
&= \{
x \mid (x \in A \vee x \in B) \vee x \in C
\} \\
&= \{
x \mid x \in (A \cup B) \vee x \in C
\} \\
&= 
(A \cup B) \cup C
\end{align*}

#### Demonstração
\begin{align*}
A \cap (B \cap C) &= \{
x \mid x \in A \wedge x \in (B \cap C)
\} \\
&= \{
x \mid x \in A \wedge (x \in B \wedge x \in C)
\} \\
&= \{
x \mid (x \in A \wedge x \in B) \wedge x \in C
\} \\
&= \{
x \mid x \in (A \cap B) \wedge x \in C
\} \\
&= 
(A \cap B) \cap C
\end{align*}

### Leis de Morgan
<ul>
    <li> $(A \cup B)^c =A^c \cap B^c$ </li>
<br/>
    <li> $(A \cap B)^c =A^c \cup B^c$ </li>
</ul>

#### Demonstração
\begin{align*}
(A \cup B)^c &= \{ x \mid x \not\in (A \cup B) \} \\
             &= \{ x \mid \neg ( x \in (A \cup B))  \} \\
             &= \{x \mid \neg (x \in A \vee x \in B)   \} \\
             &= \{x \mid \neg (x \in A) \wedge \neg (x \in B)  \} \\
             &= \{x \mid x \not\in A \wedge x \not\in B  \} \\
             &= \{x \mid x \in A^c \wedge x \in B^c   \} \\
             &= A^c \cap B^c
\end{align*}

#### Demonstração
\begin{align*}
(A \cap B)^c &= \{ x \mid x \not\in (A \cap B) \} \\
             &= \{ x \mid \neg ( x \in (A \cap B))  \} \\
             &= \{x \mid \neg (x \in A \wedge x \in B)   \} \\
             &= \{x \mid \neg (x \in A) \vee \neg (x \in B)  \} \\
             &= \{x \mid x \not\in A \vee x \not\in B  \} \\
             &= \{x \mid x \in A^c \vee x \in B^c   \} \\
             &= A^c \cup B^c \quad \blacksquare
\end{align*}

### Distributivas
<ul>
    <li> $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$ </li>
<br/>
    <li> $A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$ </li>
</ul>

#### Demonstração
\begin{align*}
A \cap (B \cup C) &= \{x \mid x \in A \wedge (x \in (B \cup C))  \} \\
&= \{x \mid x \in A \wedge (x \in B \vee x \in C)  \} \\
&= \{x \mid (x \in A \wedge x \in B) \vee (x \in A \wedge x \in C) \} \\
&= \{x \mid (x \in (A \cap B)) \vee (x \in (A \cap C))  \} \\
&= (A \cap B) \cup (A \cap C) \quad \blacksquare
\end{align*}

#### Demonstração
\begin{align*}
A \cup (B \cap C) &= \{x \mid x \in A \vee (x \in (B \cap C))  \} \\
&= \{x \mid x \in A \vee (x \in B \wedge x \in C)  \} \\
&= \{x \mid (x \in A \vee x \in B) \wedge (x \in A \vee x \in C) \} \\
&= \{x \mid (x \in (A \cup B)) \wedge (x \in (A \cup C))  \} \\
&= (A \cup B) \cap (A \cup C) \quad \blacksquare
\end{align*}

### Absorção
<ul>
    <li> $A \cup (A \cap B) =A$ </li>
<br/>
    <li> $A \cap (A \cup B) =A$ </li>
</ul>

#### Demonstração
\begin{align*}
i) \quad & A \cap B  \subset A \\
         & A \cup (A \cap B) \subset A \cup A \\
         & A \cup (A \cap B) \subset A \\
         & \\
ii) \quad & A \subset A \cup (A \cap B) 
\end{align*}
<br/>
<p> De i) e ii) temos que $A \cup (A \cap B) =A. \quad \blacksquare $ </p>

#### Demonstração
\begin{align*}
A \cap (A \cup B) &= (A \cap A) \cup (A \cap B) \\
&= A \cup (A \cap B) \\
&= A \quad \blacksquare
\end{align*}

<br/>
    <li> $A \cap A^c  = \emptyset$ </li>
</ul>

### Solução de $\; a)$
<p> Pelo absurdo: vamos supor que $\sqrt{pq}$ é racional: $\sqrt{pq}=\dfrac{m}{n}$ com $m, \; n \in \mathbb{Z}$ primos entre si $(n \neq 0)$. </p>
\begin{align}
\sqrt{pq}=\dfrac{m}{n} \Rightarrow & pq = \dfrac{m^2}{n^2} \Rightarrow m^2 = n^2 p q \qquad  (1) \\
                       \Rightarrow & p|m \wedge q|m \qquad (2)\\
                       \Rightarrow & p^2|m^2 \wedge q^2|m^2 \Rightarrow m^2=p^2 q^2 r \qquad (3) \\
          (3)=(1)      \Rightarrow & n^2 p q = p^2 q^2 r \Rightarrow n^2 =pqr \Rightarrow p|n \wedge q|n \qquad (4)
\end{align}
<p> de (2) e (4), temos que $p$ e $q$ são divisores de $m$ e $n$, o qual é um absurdo, pois $m$ e $n$ são primos entre si. </p>
<br/>
A contradição vem do fato de supor que $\sqrt{pq}$ é racional. Portanto, comcluimos que $\sqrt{pq}$ é irracional. $\blacksquare$

### Solução de $\; b)$
<p> Pelo absurdo: vamos supor que $\sqrt{p}+\sqrt{q}$ é racional: $\sqrt{p}+\sqrt{q}=\dfrac{m}{n}$ com $m, \; n \in \mathbb{Z}$ primos entre si $(n \neq 0)$. </p>
\begin{align}
\sqrt{p}+\sqrt{q}=\dfrac{m}{n} \Rightarrow & \left( \sqrt{p}+\sqrt{q} \right)^2 =\dfrac{m^2}{n^2} \\
                               \Rightarrow & p + q + 2 \sqrt{pq} = \dfrac{m^2}{n^2} \\
                               \Rightarrow & \sqrt{pq} = \underbrace{\dfrac{1}{2}}_{\in Q} 
                               \left( \underbrace{\dfrac{m^2}{n^2}}_{\in Q} -\underbrace{p}_{\in Z}-\underbrace{q}_{\in Z} \right) \in \mathbb{Q} \quad \text{contradição, pois por (a) } \sqrt{pq} \text{ é irracional}
\end{align}
<br/>
concluimos que $\sqrt{p}+\sqrt{q}$ é irracional.

### Solução de a)
<p> <b> i) $f(A \cap B) \subset f(A) \cap f(B)$ </b> já foi provado. </p>
<br/>
<p> <b> ii) $ \supset$: </b> </p>
\begin{align}
y \in f(A) \cap f(B) \Rightarrow & y \in f(A) \wedge y \in f(B) \\
                     \Rightarrow & (\exists x_1 \in A: y=f(x_1)) \wedge (\exists x_2 \in B: y=f(x_2)) \\
                     \Rightarrow & f(x_1)=f(x_2) \rightarrow x_1=x_2 \qquad \text{(por ser f injetiva)} \\
                     \Rightarrow & \text{logo, } \exists (x \in A \wedge x \in B): y = f(x) \\
                     \Rightarrow & \exists x \in A \cap B: y=f(x) \\
                     \Rightarrow & y \in f(A \cap B)
\end{align}
<p> logo, $f(A) \cap f(B) \subset f(A \cap B)) \blacksquare$ </p>

### Questão 05 - ENQ 2019-1


### Solução de a)
<p> $f:  \rightarrow B$ é bijetiva se a relação entre seus elementos é um a um, logo $a=n(A)=n(B)=b$ </p>
Se $A= \{ a_1, a_2, \cdots, a_n \}$ e $B= \{ b_1, b_2, \cdots, b_n \}$, temos
<ul>
    <li>  $f(a_1)$ pode ser qualquer dos $b$ elementos de $B$</li>
    <li>  $f(a_2)$ pode ser qualquer dos $b-1$ elementos restantes de $B$</li>
    <li>  $f(a_3)$ pode ser qualquer dos $b-2$ elementos de $B$</li>
    <li> $\cdots \cdots$ </li>
    <li>  $f(a_n)$ é o último elemento restante de $B$</li>
</ul>
logo existem $b \cdot (b-1) \cdot (b-2) \cdots 1 = b!$ funções bijetivas de $A$ em $B$.

### Solução de b)
<p> $f:  \rightarrow B$ é injetiva se $f(a_i) \neq f(a_j)$, então $a_i \neq a_j$. Logo, a relação entre seus elementos é  $a=n(A) \leq n(B)=b$ </p>
Se $A= \{ a_1, a_2, \cdots, a_n \}$ e $B= \{ b_1, b_2, \cdots, b_n \}$, temos
<ul>
    <li>  $f(a_1)$ pode ser qualquer dos $b$ elementos de $B$</li>
    <li>  $f(a_2)$ pode ser qualquer dos $b-1$ elementos restantes de $B$</li>
    <li>  $f(a_3)$ pode ser qualquer dos $b-2$ elementos de $B$</li>
    <li> $\cdots \cdots$ </li>
    <li>  $f(a_n)$ pode ser qualquer dos $b-a+1$ dos elementos restantes de $B$</li>
</ul>
logo existem $b \cdot (b-1) \cdot (b-2) \cdots (b-a+1) = \dfrac{b!}{(b-a)!}$ funções injetivas de $A$ em $B$.
